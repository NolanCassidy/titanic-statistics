```python
import pandas as pd
import numpy as np
import random as r
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
df.info()
```

```python
df['Sex'].value_counts().plot(kind='bar')
plt.show()
```

This bar graph represents the number of females and males on the titanic. It appears that a little over 300 females and almost 600 men were on the ship. This would make sense because men were higher up economically. 

```python
pd.crosstab(index=df["Survived"],columns="count")
```

This table shows how many deaths (0) and survivors (1). It appears only 3/8 of people survived the crash. 

```python
df['Age'].hist(bins=80)
plt.show()
```

This is a histogram of all the ages of passangers. The greatest peaks are within the 18-35 range. There are also a higher amount of babies thann there are teens and elderly making me thing there may be more new families on the boat.

```python
df['Siblings/Spouses Aboard'].hist(bins=8)
plt.show()
```

This histogram shows that a majority of the pasangers have 0 spouses/siblings. The second greatest is familys consisting of only one or two others. Maybe less families on this trip.

```python
df.plot(kind='scatter',x='Age',y='Siblings/Spouses Aboard')
plt.show()
```

This scatter plot shows as age increases the pasenger is less likely to have more siblings.

```python
df.plot(kind='scatter',x='Age',y='Fare')
plt.show()
```

```python
The distribution of ticket price seems to be pretty even across all ages. There are two tickets bought for a lot more than
the average at age 40.
```

```python
df['Age'].plot.box()
plt.show()
```

This is a age box plot showing a median of about 28 with more adults than children.

```python
df.groupby(['Survived']).boxplot(column='Age', by='Sex')
plt.show();
```

This graph shows the ammount of female/male that survived and didn't. There were many more females who survived making me think they were given priority. This is also interesting due to how many more men were on the boat. 

```python
df.plot(kind='scatter',x='Pclass',y='Fare')
plt.show()
```

It appears a Pclass of 1 has a higher average fare price.

```python
df.groupby(['Survived'])['Age'].hist(bins = 80)
plt.show();
```

This is a histogram of who survived and who didn't across all ages. It appears to be pretty even for people surviving and dying proportionally to the count. Although it appears more babies survived than died making me think the tried to save them first.
